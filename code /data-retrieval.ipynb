# Get the list of all assets in the MSCI World Index

#### If this is not the first time you run this code, you can set `use_cache = True` to use the cached data. Otherwise, set it to `False` to always get the latest data.

```python
use_cache = False
```

```python
data_path = "../data/"
```

```python
if use_cache:
    msci_world = pd.read_csv(data_path + "msci-world-assets-tickers.csv")
else:
    msci_world = pd.read_csv(data_path + "msci-world-assets.csv")
    msci_world.head()
```

We  will use the `yfinance` library to retrieve the list of all assets in the MSCI World Index. The company name retrieved on the `msci-world-assets.csv` file usually differs from the one on Yahoo Finance. The following code helps to match the company names in the `msci-world-assets.csv` file with the ones on Yahoo Finance.

```python
if not use_cache:
    for i in range(len(msci_world)):
        if msci_world.loc[i, "compagny"].endswith(" A"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" A", "")
        if msci_world.loc[i, "compagny"].endswith(" B"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" B", "")
        if msci_world.loc[i, "compagny"].endswith(" C"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" C", "")
        if msci_world.loc[i, "compagny"].endswith(" GRP"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" GRP", "")
        if msci_world.loc[i, "compagny"].endswith(" GROUP"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" GROUP", "")
        if msci_world.loc[i, "compagny"].endswith(" GROUPE"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" GROUPE", "")
        if msci_world.loc[i, "compagny"].endswith(" CRP"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" CRP", "")
        if msci_world.loc[i, "compagny"].endswith(" CORP"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" CORP", "")
        if msci_world.loc[i, "compagny"].endswith(" INDS"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" INDS", "")
        if msci_world.loc[i, "compagny"].endswith(" HLDGS"):
            msci_world.loc[i, "compagny"] = msci_world["compagny"][i].replace(" HLDGS", "")
```

Now we will use the `yfinance` library to retrieve the ticker symbols for the companies in the MSCI World Index.

```python
if not use_cache:
    msci_world["ticker"] = None
    for i in range(len(msci_world)):
        try:
            msci_world.loc[i, "ticker"] = yf.Search(msci_world["compagny"].iloc[i]).all['quotes'][0]['symbol']
        except:
            print(f"Didn't found a ticker for {msci_world['compagny'].iloc[i]}")
```

```python
len(msci_world) - len(msci_world[msci_world["ticker"].isnull()]),len(msci_world)
```

We successfully retrieved 1175/1242 ticker symbols for the companies in the MSCI World Index. We will save the ticker symbols in a CSV file called `msci-world-tickers.csv`.

```python
if not use_cache:
    msci_world.to_csv(data_path + "msci-world-assets-tickers.csv", index=False)
```

Get the data for all assets in the MSCI World Index

```python
for i in range(len(msci_world)):
    # Skip if the ticker is None or NaN
    if msci_world["ticker"].iloc[i] is None or pd.isna(msci_world["ticker"].iloc[i]):
        continue
    try:
        # Retrieve the historical data for the ticker and save it to a CSV file
        print(f"Getting data for {msci_world['ticker'].iloc[i]}")
        data = yf.Ticker(msci_world["ticker"].iloc[i]).history(period="max")
        data.to_csv(data_path + "yahoo-finance-dataset/"  + msci_world["ticker"].iloc[i] + ".csv")
    except:
        # Handle the case where data retrieval fails
        print(f"Couldn't get data for {msci_world['ticker'].iloc[i]}")
```
